```python
#  
nounstype = ['NN', 'NNS', 'NNP', 'NNPS', 'PRP', 'PRP$']
nouns = []
for words in tags_en:
    for noun in nounstype:
        if noun in words:
            nouns.append(words[0])
nouns[:10]
```

```python
#  
verbtype = ['VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ']
verbs = []
for words in tags_en:
    for verb in verbtype:
        if verb in words:
            verbs.append(words[0])
verbs[:10]
```

```python
#  
adjtype = ['JJ', 'JJR', 'JJS']
adjs = []
for words in tags_en:
    for adj in adjtype:
        if adj in words:
            adjs.append(words[0])
adjs[:10]
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('hotel-reviews.csv')['Description']
```

```python
df.head()
```

```python
df = df.str.replace('[^a-zA-Z]',' ')
```

```python
df = df.str.lower()
```

```python
df.head()

df = df.str.split()
```

```python
from gensim.models.word2vec import Word2Vec
model = Word2Vec(df,size=100, window=3,workers=4, min_count=5, sg=1)

model.init_sims(replace=True)
```

```python
print(model.wv.similarity('room', 'hotel'))
```

```python
#  https://stackoverflow.com/questions/43776572/visualise-word2vec-generated-from-gensim
from sklearn.manifold import TSNE
import matplotlib as mpl
import matplotlib.pyplot as plt
import gensim 
import gensim.models as g

#       
mpl.rcParams['axes.unicode_minus'] = False

vocab = list(model.wv.vocab)
X = model[vocab]

print(len(X))
print(X[0][:10])
tsne = TSNE(n_components=2)

# 100   
X_tsne = tsne.fit_transform(X[:100,:])
# X_tsne = tsne.fit_transform(X)
```

```python
df = pd.DataFrame(X_tsne, index=vocab[:100], columns=['x', 'y'])
df.shape
```

```python
df.head(10)
```

```python
fig = plt.figure()
fig.set_size_inches(40, 20)
ax = fig.add_subplot(1, 1, 1)

ax.scatter(df['x'], df['y'])

for word, pos in df.iterrows():
    ax.annotate(word, pos, fontsize=30)
plt.show()
```
